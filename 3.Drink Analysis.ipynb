 # EDA
 Basic information of attributes
 - Country :  name of contury
 - beer_servings : consumption of beer
 - spirit_serving : consumption of spirit
 - wine_servig : consumption of wine
 - total_liters_of_pure_alchohol : total amount of alcohol consumption
 - continent : region

```python
# import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# drinks csv file located in same folder
drinks = pd.read_csv('drinks.csv')
print(drinks.info())
drinks.head(10)
```

```python
# There are twenty three missing data in continent column
# Check quantitative data
# Continent and countries are qualitative data
drinks.describe()
```

```python
# start visulazing to see the insight
# check correlation between beer and wine
# we are going to use pearson. Pearson method is one of the population method to analyze the two data
corr = drinks[['beer_servings','wine_servings']].corr(method = 'pearson')
print(corr)
```

```python
corr1 = drinks[['wine_servings','beer_servings']].corr(method = 'pearson')
print(corr)
```

```python
# correlation among all features
corr_all_feature = drinks[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']].corr(method = 'pearson')
print(corr_all_feature)
```

```python
# visualize the correlation using heatmap
# heatmap has to use correlation
import seaborn as sns
import matplotlib.pyplot as plt

col_view = ['beer','wine','spirit','alcohol']
sns.set(font_scale = 1.3)
hm = sns.heatmap(corr_all_feature.values,
                cbar = True,
                annot = True,
                square = True,
                fmt = '.2f',
                annot_kws = {'size':15},
                yticklabels = col_view,
                xticklabels = col_view)

plt.tight_layout()
plt.show()
```

```python
# visualize the correlation using pairplot
sns.set(style = 'whitegrid', context = 'notebook')
sns.pairplot(drinks[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']], height = 3)
```

```python
# fill the missing data in continent
# we have mentioned that there are twenty three missing data in continent
# We fill missing data to OT = other
drinks['continent'] = drinks['continent'].fillna('OT')
drinks.head(10)
```

```python
# ratio of other continent using pie-chart
# continent name into list
labels = drinks['continent'].value_counts().index.tolist()
print(labels)
# number of continent into list
fracs1 = drinks['continent'].value_counts().values.tolist()
print(fracs1)
# How far away from the origin
explode = (0,0,0,0.25,0,0)

plt.pie(fracs1, explode = explode, labels = labels, autopct = '%.0f%%', shadow = False)
plt.title('null data to \'OT\'')
plt.show()
```

```python
# agg provide basic statistic data
result = drinks.groupby('continent').spirit_servings.agg(['mean','min','max','sum'])
result.head()
```

```python
drinks.columns
```

```python
total_mean = drinks.total_litres_of_pure_alcohol.mean()
print(total_mean)
```

```python
continent_mean = drinks.groupby('continent')['total_litres_of_pure_alcohol'].mean()
print(continent_mean)
```

```python
continent_over_mean = continent_mean[continent_mean  >= total_mean]
print(continent_over_mean)
```

```python
# Which continent has the maximum beer serving?
total_mean_beer = drinks.beer_servings.mean()
print(total_mean_beer)
```

```python
beer_continent = drinks.groupby('continent').beer_servings.mean().idxmax()
print(beer_continent)
```

```python
# visualizing result spirit serving use bar graph
n_groups = len(result.index)

means = result['mean'].tolist()
mins = result['min'].tolist()
maxs = result['max'].tolist()
sums = result['sum'].tolist()

print(means)
print(mins)
print(maxs)
print(sums)


index = np.arange(n_groups)
bar_width = 0.1

rects1 = plt.bar(index, means, bar_width, color = 'r', label = 'Mean')
rects2 = plt.bar(index + bar_width, mins, bar_width, color = 'g', label = 'Min')
rects3 = plt.bar(index + bar_width * 2, maxs, bar_width, color = 'b', label = 'Max')
rects4 = plt.bar(index + bar_width * 3, mins, bar_width, color = 'y', label = 'Sum')

plt.xticks(index, result.index.tolist())
plt.legend()
plt.show()
```

```python
# Pure alcohol

continents = continent_mean.index.tolist()
continents.append('mean')
x_pos = np.arange(len(continents))
alcohol = continent_mean.tolist()
alcohol.append(total_mean)

bar_list = plt.bar(x_pos, alcohol, align = 'center', alpha = 0.5)
bar_list[len(continents)-1].set_color('g')
plt.plot([0,6], [total_mean, total_mean], "k--")
plt.xticks(x_pos, continents)

plt.ylabel('total_liters_of_pure_alcohol')
plt.title('total_liters_of_pure_alcohol by Continent')

plt.show()
```

```python
# beer serving
beer_group = drinks.groupby('continent')['beer_servings'].sum()
continents = beer_group.index.tolist()
y_pos = np.arange(len(continents))
alcohol = beer_group.tolist()

bar_list = plt.bar(y_pos, alcohol, align = 'center', alpha = 0.5)
bar_list[continents.index('EU')].set_color('g')
plt.xticks(y_pos, continents)
plt.ylabel('beer_servings')
plt.title('beer_servings by Continent')

plt.show()
```

```python
# t-test between Africa and Europe

africa = drinks.loc[drinks['continent'] == 'AF']
europe = drinks.loc[drinks['continent'] == 'EU']

#import scipy
from scipy import stats
tTestResult = stats.ttest_ind(africa['beer_servings'], europe['beer_servings'])
tTestResultDiffvar = stats.ttest_ind(africa['beer_servings'], europe['beer_servings'], equal_var = False)

print("The t-statistic and p-value assuming equal variances is %.3f and %.3f"% tTestResult)
print("The t-statistic and p-value not assuming equal variances is %.3f and %.3f"% tTestResultDiffvar)
```

```python
drinks['total_servings'] = drinks['beer_servings'] + drinks['wine_servings'] + drinks['spirit_servings'] 
```

```python
drinks['alcohol_rate'] = drinks['total_litres_of_pure_alcohol'] / drinks['total_servings']
drinks['alcohol_rate'] = drinks['alcohol_rate'].fillna(0)
```

```python
country_with_rank = drinks[['country', 'alcohol_rate']]
country_with_rank = country_with_rank.sort_values(by = ['alcohol_rate'], ascending = 0)
country_with_rank.head(10)
```

```python
country_list = country_with_rank.country.tolist()
x_pos = np.arange(len(country_list))
rank = country_with_rank.alcohol_rate.tolist()

bar_list = plt.bar(x_pos, rank)
bar_list[country_list.index('South Korea')].set_color('r')
plt.ylabel('alcohol rate')
plt.title('liquor drink rank by contry')
plt.axis([0,200,0,0.3])

korea_rank = country_list.index('South Korea')
korea_alc_rate = country_with_rank[country_with_rank['country'] == 'South Korea']['alcohol_rate'].values[0]
plt.annotate('South Korea :' + str(korea_rank - 1),
            xy = (korea_rank, korea_alc_rate),
            xytext = (korea_rank+ 10, korea_alc_rate + 0.05),
             arrowprops = dict(facecolor = 'red', shrink = 0.05))
plt.show()
```
